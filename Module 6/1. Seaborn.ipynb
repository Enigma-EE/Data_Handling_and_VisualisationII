First, let's look at the difference in the visuals and code when building your plot with Pandas versus Seaborn.

As always, start by importing the libraries.

```python
# libraries

import numpy as np # manipulation of arrays
import pandas as pd # manipulation of data as DataFrame or Series

# Libraries for matplotlib

import matplotlib.pyplot as plt
import seaborn as sns # seaborn library

%matplotlib inline
```

We start with a time series from the previous "Plotting a Pandas Series with Series.plot()" example.

```python
# create random data for our Time Series
# the seed to generate the same random numbers as in this example
seedValue = 20
np.random.seed(seedValue)

# create a range of dates that will become the index
dateRange = pd.date_range('2016-01-01', '2019-10-07')

# generate the random data values(1 for each of date in the date range)
numberGen = len(dateRange) # 1376 numbers

# generate 4 columns of random numbers
df_data = np.random.randn(numberGen,4)

# create the Time Frame(a Data Frame with a DateTimeIndex)
df_timeFrame = pd.DataFrame(df_data, index = dateRange,
                           columns = ['Random1', 'Random2', 'Random3', 'Random4'])

# apply a cumulative sum function to the data
df_sumwalk = df_timeFrame.cumsum()

# plot it
df_sumwalk.plot(color = ['b','k','r','g'],
                title='Four Accumulate random Walks from 2015-01-01 to 2019-10-07')

# set labels in x-axis and y-axis, and legend
plt.xlabel('Date')
plt.ylabel('Accumulate sum')
plt.legend(loc='lower left')
```

Now, plotting the same data with Seaborn.

```python
# turn on seaborn styling
sns.set()
```

```python
# plot it
df_sumwalk.plot(color = ['b','k','r','g'],
                title='Four Accumulate random Walks from 2015-01-01 to 2019-10-07')

# set labels in x-axis and y-axis, and legend
plt.xlabel('Date')
plt.ylabel('Accumulate sum')
plt.legend(loc='lower left')
```

Seaborn adds styling features that make the plot easier to read.
