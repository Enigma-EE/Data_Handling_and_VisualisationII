You've previously seen how to use boxplots in Pandas to show you how your data is distributed, what the minimum values and the maximum values are, and outlying data (data that doesn't follow the trend).

Seaborn produces quick, easy-to-read boxplots.

```python
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns  # seaborn library
%matplotlib inline
#Turn on seaborn styling
sns.set()

#import the data 
df = pd.read_csv('weatherAUS.csv', parse_dates=['Date'])
```

```python
#box plot showing the max temperature per location using the Seaborn library 
plt.figure(figsize=(12,14))

# boxplot just needs to know the columns you want represented and the data
sns.boxplot('MaxTemp', 'Location', data=df, );
```

In this example, you can see that MountGinini has the lowest temperature. On the other hand, Woomera has the highest temperature. The diamonds represent outliers’ values.
