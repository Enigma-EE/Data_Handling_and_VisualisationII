```python
# data from 
# https://data.gov.au/dataset/ds-sa-e0cc2850-5096-44ec-9602-7b5abfaa2094/details?q=reptiles

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('records-2021-06-04.csv')

df.head(5)
```

```python
# size
df.shape
```

```python
df1 = df.dropna(how='all', axis=1)
df1
```

```python
# Must contain dates, numbers and strings
# dates -> year float64, month float64, day float64, eventDate object, eventTime object                        object
# numbers -> float64
# strings -> object
df1.dtypes
```

## Bar Plot

```python
fs = df['family'].squeeze()
fsc = fs.value_counts().rename_axis('f_rec_counts').reset_index(name='f_counts')

ss = df['stateProvince'].squeeze()
ssc = ss.value_counts().rename_axis('s_rec_counts').reset_index(name='s_counts')
```

```python
ssc
```

```python
ssc_new = ssc.set_index('s_rec_counts')
ssc_new
```

```python
bar_plot1 = ssc_new.plot.bar(figsize=(10,8),
                          title = 'Bar plot of occurence in stateProvince')
bar_plot1.set_ylabel('counts')
```

```python
fsc
```

```python
fsc_new = fsc.set_index('f_rec_counts')
bar_plot2 = fsc_new.plot.bar(figsize=(10,8),
                          title = 'Bar plot of occurence by family')
bar_plot2.set_ylabel('counts')
```

## Scatterplot

```python

scaP = df1.plot.scatter(x='decimalLatitude', y='decimalLongitude', 
                        c='individualCount', title = 'Decimal coordinates of records', 
                        figsize=(10,7),colormap='viridis')

scaP.set_xlabel('decimal Latitude')
scaP.set_ylabel('decimalLongitude')
```

## Pie plot

```python
df1.stateProvince.value_counts().plot.pie(autopct='%1.1f%%', figsize=(20,12), 
                                          title = 'Occurence by state pie plot',startangle=90)



plt.legend(loc="upper left")
```
