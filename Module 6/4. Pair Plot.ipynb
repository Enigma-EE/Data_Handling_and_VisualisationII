A pair plot is similar to a scatter plot, but a pair plot represents the relationship between two variables. The closer the data is to a line, the more related the two variables are. This is another way of seeing how variables are related.

Seaborn will automatically generate all of the pair plots from your Pandas Data Frame: plotting each column against the other columns.

**sns.pairplot(<<Data>>)**

```python
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns  # seaborn library
%matplotlib inline
#Turn on seaborn styling
sns.set()

#import the data 
df = pd.read_csv('weatherAUS.csv', parse_dates=['Date'])
```

```python
#Using the Australia Rain Data 
#Showing the relationship between pairs of atributes for Perth 
#We filter by Perth using df[df.Location=='Perth']
#Inside the second square brackets are the names of the columns 
#that you want to know the correlations of
sns.pairplot(df[df.Location=='Perth'][['MinTemp', 'MaxTemp', 'Rainfall', 'WindGustSpeed',
                                       'Humidity9am','Humidity3pm','Pressure9am',
                                       'Pressure3pm']])
```

In the pair plots, because we are looking to have a correlation, we try to find a line. In this plot, there is an evident line in the plot between pressure at 9am and pressure at 3pm; this means that these two variables are correlated.
