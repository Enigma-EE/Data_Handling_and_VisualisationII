```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df_data = pd.read_csv("weatherAUS.csv", parse_dates=['Date'])
df_data.head()
```

```python
# I'm going to limit my data to the December 2015

df_2015 = df_data[(df_data.Date>=pd.Timestamp('12/1/2015')) & 
                (df_data.Date<=pd.Timestamp('12/31/2015'))]

#Group same dates together and find the max of the MaxTemp values for each group
df_filtered = df_2015.sort_values('Date').groupby('Date')['MaxTemp'].max()
df_filtered
plot = df_filtered.plot.barh(title="Max Temp by Date", figsize=(12,12))

plot.set_xlabel('Temp in Celcius')
```

## Practise Groupby()

Using this data set from weatherAus.csv  download, plot:

a horizontal bar plot with the maximum temperature per location
a bar plot, choosing from any of the columns except MaxTemp and MinTemp.

```python
#Solution 
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df_data = pd.read_csv("weatherAUS.csv", parse_dates=['Date'])


plot = df_data.groupby('Location')['MaxTemp'].max().plot.barh(title="Hottest Temps",figsize=(12,12), color= 'coral')
plot.set_xlabel = "Temperature in Celcius"
```
