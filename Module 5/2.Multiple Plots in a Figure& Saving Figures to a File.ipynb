Often you want to compare data and you can make this happen by making multiple calls for each plot that you want to generate.

In this example, two NumPy functions, the sine and cosine functions from trigonometry, are applied to the data. You can look at other mathematical functions at SciPy.org https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html. 

In the first plot, we only specified the y value and the x value was just the index where the value was stored in the array (0,1,2...999). In these plots, we will specify both the x and y values. The x-axis will have our array values, and the y-axis will will have value sin(x) in one plot and cos(x) in the second plot.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# create a sequence of 1000 float numbers from 0 to 10
data = np.linspace(0,10,1000)

# give this figure a name so we can refer to it later
plot2 = plt.figure()

# x values come from the data array, y value are sin(data) or cos(data)
plt.plot(data,np.sin(data)) # plot sin blue line
plt.plot(data,np.cos(data)) # plot cos orange line
```

You can save your generated figure using the savefig() function. If you don't specify a directory the figure is saved in the same directory as your Jupyter Notebook.

```python
# saving the plot
# here is where we use that name we saved earlier!
# this tells matplotlib which figure to save
plot2.savefig('plot2.png')
```
