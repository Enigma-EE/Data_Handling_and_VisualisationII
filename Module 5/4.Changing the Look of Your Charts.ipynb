You can change many of the attributes of your plots, such as:

1. colour of the line: color = <<color_name>>
2. line style: linestyle ='dashed' or linestyle = '--'
3. markers: marker = 'o'

You can find a full list of the options in the notes section at 
matplotlib https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)

# using different format to define a color and marker
plt.plot(x,np.sin(x - 0), color='blue', marker='o') # specify color by name
plt.plot(x,np.sin(x - 1), color='g', marker='<') # short color code (rgbcmyk)
plt.plot(x,np.sin(x - 2), color='0.75', marker='>') # Grayscale between 0 and 1
plt.plot(x,np.sin(x - 3), color='#FFDD44', marker='.') # Hex code (RRGGBB from 00 to FF)
plt.plot(x,np.sin(x - 4), color=(1.0,0.2,0.3), marker='*') # RGB tuple, value 0 and 1
plt.plot(x,np.sin(x - 5), color='chartreuse', marker='+') # all HTML color names supported
```

```python
# using different line style
plt.plot(x,np.sin(x-0),linestyle='solid')
plt.plot(x,np.sin(x-1),linestyle='dashed')
plt.plot(x,np.sin(x-2),linestyle='dashdot')
plt.plot(x,np.sin(x-3),linestyle='dotted')
```

```python
# using different line style
plt.plot(x,np.sin(x-0),linestyle='-') # soild
plt.plot(x,np.sin(x-1),linestyle='--') # dashed
plt.plot(x,np.sin(x-2),linestyle='-.') # dashdot
plt.plot(x,np.sin(x-3),linestyle=':') # dotted
```

```python
# you can combine the line style and color
plt.plot(x,np.sin(x-0),'-r') # soild red
plt.plot(x,np.sin(x-1),'--k') # dashed black
plt.plot(x,np.sin(x-2),'-.y') # dashdot yellow
plt.plot(x,np.sin(x-3),':g') # dotted green
```

## Pyplot styles

There are many styles that you can specify to change the appearance of all of your plots. For example, you might want to add gridlines. Gridlines can make it easier to read charts by providing a grid to show the plots, similar to graph paper.

One of the pyplot styles that has a grid is the 'seaborn-whitegrid' style. You can make pyplot use this style using plt.style.use('seaborn-whitegrid')

```python
plt.style.use('seaborn-whitegrid')

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
# you can see all of the available styles
plt.style.available
```

Try the different styles by replacing 'seaborn-whitegrid' with other options and running your code cell.

```python
plt.style.use('bmh')

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```
