Sometimes it can be more helpful to display the plots separately but still in the same figure. To do this you can use the subplots() function. You can find more information on subplots at matplotlib https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots.html

subplots() takes as an argument the number of subplots you want to include in your figure. Subplots returns two values. The first value returned is a reference to the figure, which you can use later to reference this figure (this reference is the same type as you got from figure() that you used to save your figure).

The second value returned is an array of plots for you to fill with your plots.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# example of subplots
# create some data
x = np.linspace(-10,10,100)

# plots will be a 2D array of Axes objects
# (Axes objects cotain plotting information for the plots)

numrows = 2
numcols = 2
fig_name,plots = plt.subplots(numrows,numcols)

# call plot() method on the appropriate index
plots[0,0].plot(x,np.sin(x))
plots[0,1].plot(x,np.cos(x))
plots[1,1].plot(x,np.tan(x))
plots[1,0].plot(x,np.cumsum(x))
```
