Titles, axis labels and simple legends play a crucial role in making plots easier to understand. In the following cells, we'll briefly look at the labelling of plots. You require more details about the methods for labelling the plots, at matplotlib https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

```python
# example of labeling a plot

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x = np.linspace(0,10,100)

plt.plot(x, np.sin(x),'-r',label='sin(x)') #plotting sin red line
plt.plot(x, np.cos(x),'-b',label='cos(x)') #plotting cos blue bline

# setting the title, lables and legend
plt.title('Sine and Cosine Curves')
plt.xlabel('x')
plt.ylabel('sin(x) and cos(x)')
plt.legend()
```

## Adjusting the axes limits

It can often be useful to focus on a specific part of a plot and you can do this by changing the x and y values that are included (axes limits). Matplotlib chooses default axes limits that fit the given data, but you can change them using the plt.xlim( ) and plt.ylim( ) methods.

```python
#First a standard plot
x = np.linspace(-5,5,100)

#original plot
plt.plot(x, np.sin(x),'-r')
```

```python
# this time change the x limit: -7 to 7
# and y limit -1.5, 1.5

# create a plot
x = np.linspace(-5,5,100)


#original plot
plt.plot(x, np.sin(x),'-r')

# change the axes

# previous x-axis limit as -6 to 6
# change the limit to -7 to 7
plt.xlim(-7,7)

# previous y-axis limit was -1 to 1
# change the limit to -1.5 to 1.5
plt.ylim(-1.5, 1.5)

#compare the previous plot and this new one
# note the changes to the x and y axis values
```

You can also use plt.axis() to change the range of your x and y axes:

tight will shrink your plot size to fit your data

equal will use equal range of values on both the x and y axes.

```python
# Example of tightening the axes bounds
plt.plot(x,np.sin(x),'-g')
plt.axis('tight')

# notice how the plot has 'zoom in' by excluding most of the axes ranges
# that weren't used by the data value
```

```python
# Example or 'equal' unit in x-axis and y-axis
plt.plot(x,np.sin(x),'-m')
plt.axis('equal')

# note how this has made the x and y ranges the same 
# with 0.0 being at the center of the plot
```

## Practise Labelling Plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
x = np.linspace(2,10,300)

plt.plot(x,np.power(x,2),'-y')
plt.plot(x,np.power(x,3),'-g')
plt.plot(x,np.log(x),'-c')
plt.plot(x,2**x,'-r')
plt.plot(x,np.log2(x),'-k')

plt.title('Five Curves')
plt.xlabel('x')
plt.ylabel('x^2, x^3, log(x), 2^x and log2(x)')
plt.legend('ygcrk')
```
