## Summary

Key concepts to take away from this section:

Pandas provides two new types: Series and Data Frames.

A Series is similar to a Python dictionary with just one column, but is built on NumPy to provide faster processing over large data sets.

A Data Frame is similar to a Python dictionary with multiple columns, but is built on NumPy to provide faster processing over large data sets.

You can set your series or Data Frame indexes to be any type. You have used:

Series with integer indexes (similar to a one-column dictionary) or with time indexes\
Data Frames with label indexes or with time indexes.
You can select parts of your data using the index(row) from a:

Series using slicing: **aSeries[start:stop:step]**\
Series using conditions: **aSeries[label condition]**

Data Frame\
using loc:\
**aFrame.loc[rowstartlabel:rowstoplabel:rowstep,colstartlabel:colstoplabel:colstep]**\
**aFrame.loc[label condition]**\
using iloc: \
**aFrame.iloc[rowstartindex:rowstopindex:rowstep,colstartindex:colstopindex:colstep]**\
