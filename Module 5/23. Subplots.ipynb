As in Matplotlib, you may want to show multiple plots in the same figure. You can draw multiple plots in the same figure using subplots().

We're going to use for loops to iterate over each row and column of our subplots. To refresh your knowledge of iteration, go back to Week 2: Iteration. https://myuni.adelaide.edu.au/courses/63752/pages/week-2-iteration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# create different DataFrames
# with this np.random.rand(10,3)*10, you generate a two-dimentional [3x10] numbers between 0 to 10
df1 = pd.DataFrame(np.random.rand(10,3)*10, columns=['R1', 'R2','R3'])
df2 = pd.DataFrame(np.random.rand(10,3)*10, columns=['R1', 'R2','R3'])
df3 = pd.DataFrame(np.random.rand(10,3)*10, columns=['R1', 'R2','R3'])
df4 = pd.DataFrame(np.random.rand(10,3)*10, columns=['R1', 'R2','R3'])
df5 = pd.DataFrame(np.random.rand(10,3)*10, columns=['R1', 'R2','R3'])
df6 = pd.DataFrame(np.random.rand(10,3)*10, columns=['R1', 'R2','R3'])

# we create  an array with all the DataFrames to plot
df_list = [df1, df2, df3, df4, df5, df6]

# you can define number of rows and columns for subplots
numberRow=2
numberCol=3

fig,plots = plt.subplots(numberRow, numberCol)
# variable to counter to move in the list of the frames and display each of them
counter=0

# plot each row
for r in range(numberRow):
    # plot each column in this row
    for c in range(numberCol):
        
        # choose the dataFrame for this row, column location and plot it
        # the ax attribute needs to be set to plots[row, column]
        # plots was returned from the call to plt.subplots(above)
        df_list[counter].plot(ax=plots[r,c], figsize=(20,10))
        
        # go to the next data frame
        counter = counter + 1
        
# notice a new parameter figsize=(20,10) is uesd to resize the overall size of the figure
# that the plots are drawn in
# try different values and see the effect on the  plots
```

## Practise Plotting Data Frames

```python
#DataFrame to divide 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


number = np.arange(0., 5., 0.2)
legend_labels = ['Solid','Dashed','Dotted', 'Dot-dashed', 'Points', 'Asterisk']
df_line = pd.DataFrame({0: number,
                       1: number**1.5,
                       2: number**2.0,
                       3: number**2.5,
                       4: number**3.0,
                       5: number**3.5}                       
                      )
#Create a plot with color and line style per each line
plots=df_line.plot(style= ['r-','g--','b:','m-.','k:','c:*'],title='Six different line styles',figsize=(10,5)) 
plots.legend(legend_labels,loc='upper left')
```

```python
# We create 6 DataFrames each per line style

df1 = pd.DataFrame(number, columns=['Solid'])
df2 = pd.DataFrame(number**1.5, columns=['Dashed'])
df3 = pd.DataFrame(number**2.0, columns=['Dotted'])
df4 = pd.DataFrame(number**2.5, columns=['Dot-dashed'])
df5 = pd.DataFrame(number**3.0, columns=['Points'])
df6 = pd.DataFrame(number**3.5, columns=['Asterisk'])

#We create an array with all the DataFrames to plot
df_list = [df1,df2,df3,df4,df5,df6]
styles = ['r-','g--','b:','m-.','k:','c:*']

#You can define number of rows and columns for subplots
numberRow=3
numberCol=2

fig, axes = plt.subplots(numberRow, numberCol)
#Variable to counter to move in the list of the frames and display each of them  
counter=0 
for r in range(numberRow):
    for c in range(numberCol):
        df_list[counter].plot(ax=axes[r,c], style=styles[counter],figsize=(20,10))
        counter=counter+1
```

## Summary


Pandas plots share a lot of similarities with Matplotlib.

Plots have default attributes that can be set when creating the plot including:

line colour\
line style\
marker style\
titles\


You can set x and y labels and the legends using:

set_xlabel\
set_ylabel\
legend


Like Matplotlib, plots are displayed in figures. The figures can hold one plot or multiple subplots.

The main difference to Matplotlib is you are plotting data in Series and Data Frames, instead of NumPy arrays and Pandas uses column names to label your data automatically.
