Once you have your data in a Timestamp indexed Data Frame or Series, you can use slicing, iloc and loc to select specific time indexes, just as you did with Data Frames and Series indexed with strings or numbers. Here are some examples:

```python
# create a time Series


import pandas as pd
import numpy as np

# option 2: using date range
# create a date range: ever day between Jan 1 2016 and July 10 2019
dateRange = pd.date_range('2016-01-01','2019-10-07')

# create simulated data for these dates
# use len() to work out how many data value you need (one per date in the dateRange)
numberGen = len(dateRange) # 1376 numbers

# create a time series, using the dateRange as index
timeSeries = pd.Series(np.random.randn(numberGen), index = dateRange)

print('Time series using date_range to create the index')
print(timeSeries)
```

```python
#Slice the data in October 2018 from the Series
timeSeries['Oct 1 2018':'Oct 31 2018']
```

```python
timeFrame = pd.DataFrame(timeSeries, columns = ['data'])

# Slice the data in October 2018 from the Data Frame using loc
timeFrame.loc['Oct 1 2018':'Oct 31 2018',:]
```

```python
# slice the data from the Data Frame by row position using iloc
timeFrame.iloc[20:26,:]
```
