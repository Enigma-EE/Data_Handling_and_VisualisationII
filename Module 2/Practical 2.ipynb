## Section 2 â€“ Unit and data point chooser

In Australia, blood glucose levels are reported in mmol/L instead of mg/dL. The conversion is:
mmol/L = (mg/dL) / 18

 * Please specify the preferred unit (*a*.mmol/L **/** *b*.mg/dL)

```python
import random

limit = 501
value = random.randint(60,126)
y = 0
unit = input("a.mmol/L; b.mg/dL: ")

while y < 501 and (unit != 'a' or unit != 'b'):
    value = random.randint(60,126)
    if unit == 'a':
            print("The blood glucose level is", round(value/18,2), "mmol/L")
            print('sample ', y)
         
    elif unit =='b':
            print("The blood glucose level is ", value, 'mg/dL')
            print('sample ', y)
    else :
        unit = input("Please enter a or b")
    y += 1
```
